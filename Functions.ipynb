```python
def hello_func(greeting, name):
  return '{} {}'.format(greeting, name)

print(hello_func('Hi', 'Ahmed'))
```

```python
def add_numbers(*args):
  print(args)
  return sum(args)

print(add_numbers(1, 2, 3, 4))
```

```python
def print_info(**kwargs):
  print(kwargs)

print_info(name='Ahmed', age=25, field='Artificial Intelligence')
```
